```python
def parseRSS( rss_url ):
    return feedparser.parse( rss_url )

def getHeadlines(rss_url):
    headlines = []
 
    feed = parseRSS(rss_url)
    for newsitem in feed['items']:
        headlines.append(newsitem['link'])
 
    return headlines
```

```python
allheadlines = []#contains the urls from all the 
feedurls = []
for link in rssLinks:
    newsurl = {
        'googlenews': link,
    }
    feedurls.append(newsurl)
```

```python
#dCountries,dEvents,dProvincePak,dProvinceChina,dMonths

#round 2

dCountries = {}
dEvents = {}
dProvincePak = {}
dProvinceChina = {}
dMonths = {}

for a in countries:
    dCountries[a] = int(hashlib.sha512(a.encode()).hexdigest()[0:10], 16)
    
for a in months:
    dMonths[a] = int(hashlib.sha512(a.encode()).hexdigest()[0:10], 16)
    
for a in events:
    dEvents[a] = int(hashlib.sha512(a.encode()).hexdigest()[0:10], 16)
    
for a in provincePak:
    dProvincePak[a] = int(hashlib.sha512(a.encode()).hexdigest()[0:10], 16)
    
for a in provinceChina:
    dProvinceChina[a] = int(hashlib.sha512(a.encode()).hexdigest()[0:10], 16)

dCountries,dEvents,dProvincePak,dProvinceChina,dMonths
```

```python
x=df[['Country','Year','Month']]
y=df['Event']

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0)

#predicting result using decision tree &
#predicting result using KNN, Naive Bayes
from sklearn import datasets 
from sklearn.metrics import confusion_matrix 


#1. Decision Tree -----------------------------
from sklearn.tree import DecisionTreeClassifier
dtree_model = DecisionTreeClassifier(max_depth = 2).fit(x_train, y_train)
#predicting
dtree_predictions = dtree_model.predict(x_test)


# creating a confusion matrix 
cm = confusion_matrix(y_test, dtree_predictions) 

dtree_predictions
```
